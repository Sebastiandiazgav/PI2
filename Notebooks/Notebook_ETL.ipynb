# ETL 

## Conocemos datos

```
import pandas as pd
```

### Lectura archivos csv

```
# Leemos archivo hechos.csv
hechos = pd.read_csv('hechos.csv')
```

```
#Mostramos nuesta DB
hechos
```

```
#Leemos archivo victimas.csv
victimas = pd.read_csv('victimas.csv')  
```

```
victimas
```

### Se hace un merge a los DB

Se combinan los archivos CSV por medio de la columna relacionada ID y ID_hecho, tenemos el DF llamado df_combinado 

```
#Combinamos archivos
df_combinado = pd.merge(hechos, victimas, left_on='ID', right_on='ID_hecho', how='inner') 
```

```
#Leemos archivo df_combinado, se visualiza la data
df_combinado
```

### Se eliminan columnas

Eliminar columnas que no se van  a utilizar algunas duplicadas, otras que no se necesitan.

```
#Se nombra variable, con las columnas a eliminar 
columnas_a_eliminar = ['HORA', 'Calle', 'Altura', 'Cruce', 'XY (CABA)' ,'Dirección Normalizada', 'ID_hecho', 'FECHA_y', 'AAAA_y', 'MM_y', 'DD_y', 'VICTIMA_y', 'FECHA_FALLECIMIENTO']

# Eliminar las columnas 
df_combinado = df_combinado.drop(columnas_a_eliminar, axis=1)
```

```
#Leer DF
df_combinado
```

### Cambiar tipo de datos

```
#Se valida tipo de datos
df_combinado.dtypes
```

```

# Cambiar N_VICTIMAS a tipo entero
df_combinado['N_VICTIMAS'] = df_combinado['N_VICTIMAS'].astype(int)

# Cambiar FECHA_x a tipo fecha
df_combinado['FECHA_x'] = pd.to_datetime(df_combinado['FECHA_x'], errors='coerce', format='%m/%d/%Y')

# Cambiar AAAA_x, MM_x, DD_x, HH, EDAD, comuna a tipos enteros
df_combinado['AAAA_x'] = df_combinado['AAAA_x'].astype(int)
df_combinado['MM_x'] = df_combinado['MM_x'].astype(int)
df_combinado['DD_x'] = df_combinado['DD_x'].astype(int)
df_combinado['HH'] = pd.to_numeric(df_combinado['HH'], errors='coerce').astype('Int64')  
df_combinado['COMUNA'] = df_combinado['COMUNA'].astype(int)
df_combinado['EDAD'] = pd.to_numeric(df_combinado['EDAD'], errors='coerce').astype('Int64')


# Cambiar pos x y pos y a tipos float
df_combinado['pos x'] = pd.to_numeric(df_combinado['pos x'], errors='coerce')
df_combinado['pos y'] = pd.to_numeric(df_combinado['pos y'], errors='coerce')
```

```
df_combinado
```

### Se validad valores nulos 

Se validan valores nulos o no correspondientes a los datos, se procesan debidamente para no tener problemas cuando se realice la exploracion de datos.

```
df_combinado.info()
```

```
# Mostrar los valores únicos presentes en la columna
valores_unicos = df_combinado['HH'].unique()
print(f'Valores únicos en la columna: {valores_unicos}')
```

```
# Reemplazar '<NA>' con NaN
df_combinado['HH'].replace('<NA>', pd.NA, inplace=True)

# Eliminar filas con valores <NA> en la columna 
df_combinado = df_combinado.dropna(subset=['HH'])
```

```
#Validar los valores que componen la columna 'PARTICIPANTES'
valores_unicos = df_combinado['PARTICIPANTES'].unique()
print(f'Valores únicos en la columna: {valores_unicos}')
```

```
# Lista de valores a excluir
valores_a_eliminar = ['MOTO-SD', 'SD-SD', 'AUTO-SD', 'PEATON-SD', 'SD-CARGAS', 'SD-AUTO', 'SD-MOTO', 'PASAJEROS-SD']

# Filtrar las filas que no contienen los valores a eliminar
df_combinado = df_combinado[~df_combinado['PARTICIPANTES'].isin(valores_a_eliminar)]
```

```
#Validar los valores que componen la columna 'VICTIMA_x'
valores_unicos = df_combinado['VICTIMA_x'].unique()
print(f'Valores únicos en la columna: {valores_unicos}')
```

```
#Validar los valores que componen la columna 'EDAD'
valores_unicos = df_combinado['EDAD'].unique()
print(f'Valores únicos en la columna: {valores_unicos}')
```

```
# Reemplazar '<NA>' con NaN
df_combinado['EDAD'].replace('<NA>', pd.NA, inplace=True)

# Eliminar filas con valores <NA> en la columna 
df_combinado = df_combinado.dropna(subset=['EDAD'])
```

```
#Validar los valores que componen la columna 'SEXO'
valores_unicos = df_combinado['SEXO'].unique()
print(f'Valores únicos en la columna: {valores_unicos}')
```

```
#Validar los valores que componen la columna 'ROL'
valores_unicos = df_combinado['ROL'].unique()
print(f'Valores únicos en la columna: {valores_unicos}')
```

```
# Filtrar las filas que no contienen 'SD' en la columna 'ROL'
df_combinado = df_combinado[df_combinado['ROL'] != 'SD']
```

```
#Validar los valores que componen la columna 'ACUSADO'
valores_unicos = df_combinado['ACUSADO'].unique()
print(f'Valores únicos en la columna: {valores_unicos}')
```

```
#Validar los valores que componen la columna 'pos x'
valores_unicos = df_combinado['pos x'].unique()
print(f'Valores únicos en la columna: {valores_unicos}')
```

```
# Eliminar filas con valores NaN en la columna 'pos x'
df_combinado = df_combinado.dropna(subset=['pos x'])
```

```
#Validar los valores que componen la columna 'pos y'
valores_unicos = df_combinado['pos y'].unique()
print(f'Valores únicos en la columna: {valores_unicos}')
```

```
# Guardar el DataFrame en un archivo CSV
df_combinado.to_csv('homicidio_df.csv', index=False)
```
